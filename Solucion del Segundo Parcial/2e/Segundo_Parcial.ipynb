```python
"""4. Multiplique dos matrices sparce de más de 1000 filas y columnas"""

# Importamos openCV
import cv2
from scipy.sparse import csr_matrix

# Importamos las dos imagenes
img1 = cv2.imread('/content/drive/MyDrive/20241semestre/DAT141/2do/02segundo parcial/cadena.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('/content/drive/MyDrive/20241semestre/DAT141/2do/02segundo parcial/sacapuntas.jpg', cv2.IMREAD_GRAYSCALE)


# ajustar la imagen dos para que sea de las mismas dimesiones que la imagen 1
img2_modificada = cv2.resize(img2, (img1.shape[0], img1.shape[1]))

# Convinamos las dos matrices
m1 = np.array(img1)
m2 = np.array(img2_modificada)

# Convertimos las matrices a matrices sparce
csr1 = csr_matrix(m1)
csr2 = csr_matrix(m2)

# Multiplicamos las matrices sparce
mul = csr1.dot(csr2)


print(mul)

```
